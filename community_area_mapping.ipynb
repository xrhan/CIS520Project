# Community Area to Coordinates
Maps the 77 community areas to their center coordinates, note the 41 has been deducted from latitude and 87 has been deducted from longitude

```python
import csv
import pandas as pd
import re
```

```python
comm_df = pd.read_csv("community_area.csv", sep = ",")
```

```python
comm_df = comm_df[["Area Number", "Coordinates"]]
```

```python
def parse_coord(x):
    lat, lon = x.split(' ')
    lat_deg, lat_min, lat_sec, _ = re.split('d|m|s', lat)
    lon_deg, lon_min, lon_sec, _ = re.split('d|m|s', lon)
    lat_total = float(lat_deg) + float(lat_min)/60 + float(lat_sec)/3600
    lon_total = float(lon_deg) + float(lon_min)/60 + float(lon_sec)/3600
    return round(lat_total-41,4), round(lon_total-87,4)
```

```python
for i in range(0, 77):
    lat, lon = comm_df["Coordinates"].map(lambda x: parse_coord(x)).iloc[i]
    comm_df.loc[i, "lat"] = lat
    comm_df.loc[i, "lon"] = lon
```

```python
comm_df = comm_df[["Area Number", "lat", "lon"]]
```

```python
comm_df.to_csv("community_mapping.csv", index = False)
```
