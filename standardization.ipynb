```python
import csv
import math
import pandas as pd
from datetime import datetime
```

```python
humidity = pd.read_csv('humidity.csv')
humidity = humidity[['datetime','Chicago']]
humidity = humidity.interpolate(method="linear")
humidity.to_csv("humidity.csv", index = False)
```

```python
taxi_trip = pd.read_csv("2014ss.csv", sep = ",")
temperature = pd.read_csv("temperature.csv", sep = ",")
temperature.rename(columns={'Chicago':'temperature'}, inplace=True)
temperature = temperature.interpolate(method="linear")
wind_speed = pd.read_csv("wind_speed.csv", sep = ",")
wind_speed.rename(columns={'Chicago':'wind_speed'}, inplace=True)
humidity = pd.read_csv('humidity.csv')
humidity.rename(columns={'Chicago':'humidity'}, inplace=True)

temperature["datetime"] = pd.to_datetime(temperature["datetime"], \
                                                 format='%m/%d/%y %H:%M').dt.floor('H');
temperature["datetime"] = temperature["datetime"].dt.strftime('%Y/%m/%d %H')
wind_speed["datetime"] = pd.to_datetime(wind_speed["datetime"], \
                                                 format='%m/%d/%y %H:%M').dt.floor('H');
wind_speed["datetime"] = wind_speed["datetime"].dt.strftime('%Y/%m/%d %H')
humidity["datetime"] = pd.to_datetime(humidity["datetime"], \
                                                 format='%Y-%m-%d %H:%M:%S').dt.floor('H');
humidity["datetime"] = humidity["datetime"].dt.strftime('%Y/%m/%d %H')

# Join with taxi dataset by timestamp
taxi_with_weather = pd.merge(taxi_trip, temperature, how='left', left_on='Trip Start Timestamp', 
                                    right_on='datetime')
taxi_with_weather = pd.merge(taxi_with_weather, wind_speed, how='left', left_on='Trip Start Timestamp', 
                                    right_on='datetime')
taxi_with_weather = pd.merge(taxi_with_weather, humidity, how='left', left_on='Trip Start Timestamp', 
                                    right_on='datetime')

taxi_with_weather = taxi_with_weather.drop(columns=['datetime_x', 'datetime_y'])

# Check if contains NaN
taxi_with_weather.isnull().values.any()

#taxi_with_weather.to_csv("14taxi_with_weather.csv", index = False)
```

```python
taxi_with_weather["dayofyear_sin"] = taxi_with_weather["dayofyear"].map(lambda x: round(math.sin(float(x)/365*2*math.pi),3));
taxi_with_weather["dayofyear_cos"] = taxi_with_weather["dayofyear"].map(lambda x: round(math.cos(float(x)/365*2*math.pi),3));

taxi_with_weather["hour_sin"] = taxi_with_weather["hour"].map(lambda x: round(math.sin(float(x)/24*2*math.pi),2));
taxi_with_weather["hour_cos"] = taxi_with_weather["hour"].map(lambda x: round(math.cos(float(x)/24*2*math.pi),2));
```

```python
# standardization
taxi_with_weather = pd.read_csv('download/13taxi_with_weather.csv')
print('temp_mean',round(taxi_with_weather['temperature'].mean(),5))
print('wind_speed_mean',round(taxi_with_weather['wind_speed'].mean(),5))
print('humidity_mean',round(taxi_with_weather['humidity'].mean(),5))
print('temp_std',round(taxi_with_weather['temperature'].std(),5))
print('wind_speed_std',round(taxi_with_weather['wind_speed'].std(),5))
print('humidity_std',round(taxi_with_weather['humidity'].std(),5))
taxi_with_weather['temperature'] = round((taxi_with_weather['temperature']-taxi_with_weather['temperature'].mean())/taxi_with_weather['temperature'].std(),3)
taxi_with_weather['wind_speed'] = round((taxi_with_weather['wind_speed']-taxi_with_weather['wind_speed'].mean())/taxi_with_weather['wind_speed'].std(),2)
taxi_with_weather['humidity'] = round((taxi_with_weather['humidity']-taxi_with_weather['humidity'].mean())/taxi_with_weather['humidity'].std(),3)

taxi_with_weather['dayofyear'] = round((taxi_with_weather['dayofyear']-182.5)/182.5,3)
taxi_with_weather['hour'] = round((taxi_with_weather['hour']-11.5)/11.5,2)
# considering transformation before, a total of 41.8 has been subtracted from lat
# a total of 87.7 has been subtracted from lon, both are then scaled to 10 times
taxi_with_weather['lat'] = (taxi_with_weather['lat'] - 0.8)*10
taxi_with_weather['lon'] = (taxi_with_weather['lon'] - 0.7)*10
taxi_with_weather.drop(columns=['datetime'], inplace = True)

taxi_with_weather.to_csv("download/13taxi_with_weather_std.csv", index = False)
```
